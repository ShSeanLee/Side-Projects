```python
import os

file_path = './'
file_names = os.listdir(file_path)
```

```python
# a, b의 설정을 변경하여 표시할 범위 설정
a = 0
b = 10
```

```python
# 1. 해당 이미지를 표시
from PIL import Image
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (15.0, 8.0)

rows = 2
columns = 5

for i in range(a, b) : 
    image = Image.open(f"{i}.gif")
    image_index = (i+1-a)     # image index 
    ttitle = f"Image{image_index}" # image title
    plt.subplot(rows, columns, image_index) # subplot 
    plt.title(ttitle)   # title 
    # // plt.axis('off')
    plt.xticks([])  # x = None 
    plt.yticks([])  # y = None
    plt.imshow(image)  
    image.close()
plt.show()
```

```python
# 2. 1에서 표시된 이미지를 보고 라벨링 작업 진행
# 자동 파일명 변경 .gif -> {각 label}.png
for i in range(a, b):
    src = os.path.join(file_path, f"{i}.gif")
    dst = input() + '.png'
    dst = os.path.join(file_path, dst)
    os.rename(src, dst)
```
